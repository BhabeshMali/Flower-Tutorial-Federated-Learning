# **Step 2**

## The installation of the **Flower Framework** may necessitate a Python version of 3.8 or higher.

## Let us see **step-by-step** procedure to install different Python versions and to switch them as per requirements

## **Step 2.a**

### Open Terminal using the Keywords **Ctrl+Alt+T** and check the current Python version using the command:

Screenshot from 2023-06-17 18-49-51.png

```python
python --version
```

## **Step 2.b**

### The legacy Python versions, including 3.9, 3.8, 3.7, and earlier, are not obtainable for installation via the default system repository of either Ubuntu 22.04 LTS Jammy JellyFish or 20.04 Focal Fossa. Consequently, it becomes imperative to incorporate the PPA provided by the esteemed "deadsnakes" team in order to effortlessly access the archived Python versions of the past. Therefore, run the following command

```python
sudo apt install software-properties-common
```

```python
sudo add-apt-repository ppa:deadsnakes/ppa
```

## **Step 2.c**

### Now, to install the required python version used the following commands as

### ***Python 3.9***

```python
sudo apt install python3.9
```

### ***Python 3.8***

```python
sudo apt install python3.8
```

### ***Python3.7*** and so on

```python
sudo apt install python3.8
```

```python
sudo apt install python3.x  # use x as per requirements
```

## **Step 2.d**

### Now, you can check what are the different Python versions available on your system using the command below

Screenshot from 2023-06-17 18-50-54.png

```python
ls /usr/bin/python*
```

### To establish a specific Python version as the default, you may utilize the provided command below to switch the version.

---

### First, assign a number to each version using a sequential number strating from 1



```python
sudo update-alternatives --install /usr/bin/python python /usr/bin/python3.10 1
```

```python
sudo update-alternatives --install /usr/bin/python python /usr/bin/python3.9 2
```

```python
sudo update-alternatives --install /usr/bin/python python /usr/bin/python3.8 3
```

and so on **x** and **y** repesents python version and a sequential integer assigned

```python
sudo update-alternatives --install /usr/bin/python python /usr/bin/python3.x y
```

### Python version switching

On this occasion, you will encounter various alternatives, encompassing Python 3.9 and 3.8 among them. Enter the corresponding numerical value for the desired Python version that you intend to establish as the system's default. For instance, assuming Python 3.9 occupies the third position, we input the numeral '3' followed by pressing the Enter key.

Screenshot from 2023-06-17 18-46-21.png

```python
sudo update-alternatives --config python
```

# **Step 3**

### Before, installing installing Flower, it is better to create a virtual environment and install the flower framework requirements

### Let us see how we can create a virtual environment using virtualenv

## **Step 3.a**

### Install the **virtualenv package**

```python
pip install virtualenv
```

## **Step 3.b**

Now, that the package has been installed, we can proceed to create a virtual environment for Python 3.8 version or any other versions.

Screenshot from 2023-06-17 19-43-20.png

```python
 virtualenv -p /usr/bin/python3.8 env
```

## **Step 3.c**

Now, activate the virutal environment **env** using the command below.

Screenshot from 2023-06-17 19-43-51.png

```python
source env/bin/activate
```

## **Step3.d**

### To deactivate the environment, just type the command

```python
deactivate
```
